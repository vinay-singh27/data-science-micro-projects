```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Distribution Plots

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#check distribution
sns.displot(tips['total_bill'],kde=False,bins=30)
```

```python
#check relation between two variables
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
#can also change the kind of plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
#pair plot
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
#kde plot
sns.kdeplot(tips['total_bill'])
```

# Categorical Plots

```python
#bar plot
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
#default estimatior in bar plot is mean. It can also be changed to other measures
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
#count plot
sns.countplot(x='sex',data=tips)
```

```python
#box plot
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
# Can do entire dataframe with orient='h'
sns.boxplot(data=tips,orient='h')
```

```python
#can also hue in box plot
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
```

```python
#violin plot
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
#strip plot
sns.stripplot(x="day", y="total_bill", data=tips)
```

# Heatmap 

```python
sns.heatmap(tips.corr())
```

```python
#cluster map
sns.clustermap(tips.corr())
```

# Grid Plot

```python
#create a pair grid object and then map it to any plot
# Map to upper,lower, and diagonal
g = sns.PairGrid(tips)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
#facet grid
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```
